## 11. ICD-9 Diagnosis Code Mapping (`diag_1`, `diag_2`)

The raw `diag_1` and `diag_2` columns contain ~700 unique ICD-9 codes each.  
We map them into **9 broad clinical categories** so they can be encoded as categorical features.

| ICD-9 Range | Category |
|---|---|
| 390–459, 785 | Circulatory |
| 460–519, 786 | Respiratory |
| 520–579, 787 | Digestive |
| 250.xx | Diabetes |
| 800–999 | Injury |
| 710–739 | Musculoskeletal |
| 580–629, 788 | Genitourinary |
| 140–239 | Neoplasms |
| Everything else | Other |

```python
def map_diagnosis(code):
    """Map a single ICD-9 code (string) to a clinical category."""
    if code == '?' or pd.isna(code):
        return 'Other'
    
    # Handle E and V codes
    if str(code).startswith('E') or str(code).startswith('V'):
        return 'Other'
    
    try:
        numeric = float(code)
    except ValueError:
        return 'Other'
    
    # Diabetes
    if 250 <= numeric < 251:
        return 'Diabetes'
    # Circulatory
    elif (390 <= numeric <= 459) or numeric == 785:
        return 'Circulatory'
    # Respiratory
    elif (460 <= numeric <= 519) or numeric == 786:
        return 'Respiratory'
    # Digestive
    elif (520 <= numeric <= 579) or numeric == 787:
        return 'Digestive'
    # Injury
    elif 800 <= numeric <= 999:
        return 'Injury'
    # Musculoskeletal
    elif 710 <= numeric <= 739:
        return 'Musculoskeletal'
    # Genitourinary
    elif (580 <= numeric <= 629) or numeric == 788:
        return 'Genitourinary'
    # Neoplasms
    elif 140 <= numeric <= 239:
        return 'Neoplasms'
    else:
        return 'Other'

print('Mapping function defined.')
```

```python
# Apply mapping to both diagnosis columns
df_selected['diag_1'] = df_selected['diag_1'].apply(map_diagnosis)
df_selected['diag_2'] = df_selected['diag_2'].apply(map_diagnosis)

print('Mapping applied successfully.\n')

print('=== diag_1 (mapped) ===')
print(df_selected['diag_1'].value_counts())
print()

print('=== diag_2 (mapped) ===')
print(df_selected['diag_2'].value_counts())
```

```python
# Verify: confirm both columns now have a small set of clean categories
print(f'diag_1 unique categories: {df_selected["diag_1"].nunique()}  ->  {sorted(df_selected["diag_1"].unique())}')
print(f'diag_2 unique categories: {df_selected["diag_2"].nunique()}  ->  {sorted(df_selected["diag_2"].unique())}')
```

## 12. Encode Categorical Features

**Encoding strategy:**

| Encoding | Features | Reason |
|---|---|---|
| **Label (binary)** | `diabetesMed`, `gender` | Binary Yes/No or Male/Female |
| **Label (ordinal)** | `insulin`, `metformin`, `max_glu_serum`, `A1Cresult`, `age` | Natural order exists |
| **One-Hot** | `race`, `diag_1`, `diag_2` | Nominal — no natural order |
| **Binary (target)** | `readmitted` | 0 = NO or >30 days, 1 = <30 days |

```python
# ---- 12a. Binary Encoding ----

# diabetesMed: No=0, Yes=1
df_selected['diabetesMed'] = df_selected['diabetesMed'].map({'No': 0, 'Yes': 1})

# gender: Female=0, Male=1  (drop 3 Unknown/Invalid rows)
df_selected = df_selected[df_selected['gender'] != 'Unknown/Invalid']
df_selected['gender'] = df_selected['gender'].map({'Female': 0, 'Male': 1})

print('Binary encoding applied:')
print(f"  diabetesMed -> {sorted(df_selected['diabetesMed'].unique())}")
print(f"  gender      -> {sorted(df_selected['gender'].unique())}")
print(f'  Rows after dropping Unknown/Invalid gender: {len(df_selected)}')
```

```python
# ---- 12b. Ordinal Encoding ----

# insulin: No=0, Down=1, Steady=2, Up=3
insulin_map = {'No': 0, 'Down': 1, 'Steady': 2, 'Up': 3}
df_selected['insulin'] = df_selected['insulin'].map(insulin_map)

# metformin: No=0, Down=1, Steady=2, Up=3
metformin_map = {'No': 0, 'Down': 1, 'Steady': 2, 'Up': 3}
df_selected['metformin'] = df_selected['metformin'].map(metformin_map)

# max_glu_serum: NaN=0 (None/not measured), Norm=1, >200=2, >300=3
glu_map = {'Norm': 1, '>200': 2, '>300': 3}
df_selected['max_glu_serum'] = df_selected['max_glu_serum'].map(glu_map).fillna(0).astype(int)

# A1Cresult: NaN=0 (not measured), Norm=1, >7=2, >8=3
a1c_map = {'Norm': 1, '>7': 2, '>8': 3}
df_selected['A1Cresult'] = df_selected['A1Cresult'].map(a1c_map).fillna(0).astype(int)

# age: ordinal bins  [0-10)=0 ... [90-100)=9
age_map = {
    '[0-10)': 0, '[10-20)': 1, '[20-30)': 2, '[30-40)': 3, '[40-50)': 4,
    '[50-60)': 5, '[60-70)': 6, '[70-80)': 7, '[80-90)': 8, '[90-100)': 9
}
df_selected['age'] = df_selected['age'].map(age_map)

print('Ordinal encoding applied:')
for col, m in [('insulin', insulin_map), ('metformin', metformin_map),
               ('max_glu_serum', glu_map), ('A1Cresult', a1c_map), ('age', age_map)]:
    print(f'  {col:20s} -> {sorted(df_selected[col].unique())}')
```

```python
# ---- 12c. One-Hot Encoding (nominal features) ----
# This REPLACES the original race, diag_1, diag_2 columns with dummy columns.
# The old columns are automatically dropped by pd.get_dummies.

nominal_features = ['race', 'diag_1', 'diag_2']

print(f'Shape BEFORE one-hot encoding: {df_selected.shape}')

df_selected = pd.get_dummies(df_selected, columns=nominal_features, drop_first=False)

# Convert bool columns to int (0/1) for consistency
bool_cols = df_selected.select_dtypes(include='bool').columns
df_selected[bool_cols] = df_selected[bool_cols].astype(int)

# Drop 'race_?' column — unknown/missing race is not a useful feature
df_selected.drop(columns=['race_?'], inplace=True)
print('Dropped column: race_?')

print(f'Shape AFTER  one-hot encoding: {df_selected.shape}')
print(f'\nNew one-hot columns:')
ohe_cols = [c for c in df_selected.columns if any(c.startswith(f'{nf}_') for nf in nominal_features)]
for col in sorted(ohe_cols):
    print(f'  - {col}')
```

```python
# ---- 12d. Target Encoding (Binary) ----
# 0 = NO readmission or readmitted AFTER 30 days  (NO, >30)
# 1 = Readmitted WITHIN 30 days                   (<30)

target_map = {'NO': 0, '>30': 0, '<30': 1}
df_selected['readmitted'] = df_selected['readmitted'].map(target_map)

print('Binary target encoding applied:')
print(f'  NO  -> 0')
print(f'  >30 -> 0')
print(f'  <30 -> 1')
print()
print(df_selected['readmitted'].value_counts().sort_index())
print()
print(df_selected['readmitted'].value_counts(normalize=True).sort_index().map('{:.2%}'.format))
```
